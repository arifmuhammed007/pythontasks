```python
import numpy as np

# Given sales data

months = np.array(["Jan", "Feb", "Mar", "Apr", "May", "Jun",

"Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])



revenue = np.array([12000, 15000, 18000, 21000, 25000, 23000,

28000, 30000, 27000, 22000, 19000, 17000])



customers = np.array([120, 135, 160, 180, 200, 190, 230, 250, 240, 210, 175, 160])



ad_spend = np.array([2000, 2200, 2500, 2800, 3200, 3100, 3500, 4000, 3700, 3300, 2900, 2700])

# basic Analysis
total_annual_revenue = revenue.sum()
avg_revenue = revenue.mean()
month_max_revenue = months[np.argmax(revenue)]
month_min_revenue = months[np.argmin(revenue)]

# customer Analysis

total_customer = customers.sum()
revenue_per_customer = revenue / customers
month_max_coustomer = months[np.argmax(customers)]

# advertisment perfomance
roi = revenue / ad_spend
month_max_roi = months[np.argmax(roi)]
correlation = np.corrcoef(ad_spend,revenue)[0,1]

# trend analysis
revenue_growth = np.zeros_like(revenue, dtype=float)
revenue_growth[1:] = ((revenue[1:] - revenue[:-1]) / revenue[:-1]) * 100
decline_months = months[1:][revenue[1:] < revenue[:-1]]


#outputs

print("TASK 1: BASIC ANALYSIS")
print("Total annual revenue:", total_annual_revenue)
print("Average monthly revenue:", round(avg_revenue, 2))
print("Month with highest revenue:", month_max_revenue)
print("Month with lowest revenue:", month_min_revenue)

print("\nTASK 2: CUSTOMER ANALYSIS")
print("Total customers in the year:", total_customer)
print("Average revenue per customer per month:", np.round(revenue_per_customer, 2))
print("Month with highest number of customers:", month_max_coustomer)

print("\nTASK 3: ADVERTISING PERFORMANCE")
print("Revenue-to-Ad Spend Ratio per month:", np.round(roi, 2))
print("Month with highest advertising ROI:", month_max_roi)
print("Correlation between ad spend and revenue:", round(correlation, 3))

print("\nTASK 4: TREND ANALYSIS")
print("Monthly revenue growth rate (%):", np.round(revenue_growth, 2))
print("Months with revenue decline compared to previous month:", decline_months)








```
