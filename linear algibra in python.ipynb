Question 4: Anomaly Detection - Outlier Scores (Vector Operations - Dot Product)
You have customer spending habits, represented by two features: average_transaction_value and number_of_transactions. customer_data = np.array([[50, 10], [60, 12], [200, 5], [55, 11], [45, 9]]) You want to assign an "anomaly score" to each customer. A simple way is to take the dot product with a "weight vector" that emphasizes unusual behavior. Let weights = np.array([0.2, -0.8]) (where a positive weight for average_transaction_value and negative for number_of_transactions might indicate high-value, low-frequency buyers as potentially anomalous).
a. Calculate the anomaly score for each customer using the dot product. b. Identify the customer with the highest anomaly score. What might this score suggest about this customer's behavior?

```python
customer_data = np.array([[50, 10], [60, 12], [200, 5], [55, 11], [45, 9]])
weights = np.array([0.2, -0.8])
```

```python
#)a. Calculate the anomaly score for each customer using the dot product
anomaly_score = customer_data @ weights
print(anomaly_score)
```

```python
#b. Identify the customer with the highest anomaly score. What might this score suggest about this customer's behavior?
most_anomalous_index = np.argmax(anomaly_score)
print(most_anomalous_index)
```

Question 5: Feature Combination Visualization (Vector Operations - Cross Product)
Imagine you have two features, x and y, representing attributes of a data point, and you want to understand how they "interact" to form a third derived feature z. While not a direct application of cross-product in data features, visualize how two 3D vectors combine using the cross-product to conceptually represent orthogonal relationships.
Consider two feature vectors in a 3D space: feature1 = np.array([1, 0, 0]) and feature2 = np.array([0, 1, 0]).


```python
feature1 = np.array([1, 0, 0])
feature2 = np.array([0, 1, 0])
z = np.cross(feature1,feature2)

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, feature1[0], feature1[1], feature1[2], color='r', label='Feature 1')
ax.quiver(0, 0, 0, feature2[0], feature2[1], feature2[2], color='g', label='Feature 2')
ax.quiver(0, 0, 0, z[0], z[1], z[2], color='b', label='Cross Product')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()

plt.show()
```

Question 8: Resource Allocation Optimization (Solving Linear Systems)
A manufacturing company produces three types of products (P1, P2, P3). Each product requires different amounts of raw materials (M1, M2, M3).
P1: 2 units of M1, 1 unit of M2, 1 unit of M3
P2: 1 unit of M1, 3 units of M2, 2 units of M3
P3: 1 unit of M1, 2 units of M2, 1 unit of M3
You have a total of 100 units of M1, 150 units of M2, and 80 units of M3 available.
a. Set up a system of linear equations to determine how many units of each product (P1, P2, P3) can be produced using all available raw materials. Represent this as Ax = b. b. Solve this system using numpy.linalg.solve(). c. Interpret the solution. What does it tell the company about its production capacity given the raw material constraints? Consider what might happen if the system were singular or inconsistent 



```python
#a. Set up a system of linear equations to determine how many units of each product (P1, P2, P3) can be produced using all available raw materials. 
A = np.array([[2, 1, 1],
              [1, 3, 2],
              [1, 2, 1]])

B = np.array([100, 150, 80])
```

```python
# b. Solve this system using numpy.linalg.solve().
x = np.linalg.solve(A,B)
print(x)
```
